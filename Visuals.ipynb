```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.read_csv("user_statistics.csv")
```

```python
df = pd.read_csv("user_statistics.csv")
```

## Visualized result in Maplotlib

```python
fig, ax1 = plt.subplots(figsize=(15, 5))
df.set_index('Username').plot.bar(ax= ax1, color=['blue', 'green'])
info = df["INFO"]
error = df["ERROR"]
for index, value in enumerate(info):
    plt.text(index,value, str(value),ha='right')
for index, value in enumerate(error):
    plt.text(index,value, str(value),ha='left')
b = plt.xticks(rotation=45)
```

```python
df1 = pd.melt(df, id_vars='Username',var_name='Status', value_name='Count')
```

```python
df1
```

## Visualized result in Seaborn

```python
fig, ax = plt.subplots(figsize=(20, 5))
ax = seaborn.barplot(x="Username", y="Count", hue="Status", data=df1)
for container in ax.containers:
    ax.bar_label(container)
b = plt.xticks(rotation=45)
```
